```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/06-Histograms/data.csv'
df = pd.read_csv(url)
df.head()
```

```python
bins = np.arange(10, 110, 10)

median_age = df['Age'].median()
```

```python
plt.style.use('fivethirtyeight')

plt.title('Age of Respondents')
plt.xlabel('Age groups')
plt.ylabel('Total Respondents')

plt.hist(df['Age'], bins=bins, edgecolor='black', log=True)
plt.axvline(median_age, color='red', linewidth='1', label='Median Age')

plt.legend()
plt.tight_layout()
```
