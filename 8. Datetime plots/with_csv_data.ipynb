```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt, dates as mpl_dates
```

```python
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv'
df = pd.read_csv(url, usecols=['Date', 'Close'])
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].dtypes
```

```python
plt.style.use('seaborn-v0_8')

plt.plot_date(x=df['Date'], y=df['Close'], linestyle='solid')

plt.gcf().set_size_inches(13.5, 7)
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
```
