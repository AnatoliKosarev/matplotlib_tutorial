```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv'
raw_data = pd.read_csv(url, index_col='Responder_id')['LanguagesWorkedWith'].str.split(';')
raw_data

res = pd.DataFrame(raw_data.to_list()).stack().value_counts()
res.head()
```

```python
plt.style.use('fivethirtyeight')

plt.tight_layout()

explode = [0, 0, 0, 0.1, 0]

res.head(5).plot.pie(title='Most popular programming languages', 
                     autopct='%1.1f%%', 
                     explode=explode, 
                     shadow=True, 
                     startangle=90, 
                     wedgeprops={'edgecolor': 'grey'})
```
