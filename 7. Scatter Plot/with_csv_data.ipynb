```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv'
df = pd.read_csv(url, names=['View Count', 'Total Likes', 'Like/Dislike Ratio'], header=0)
df.info()
```

```python
plt.style.use('seaborn-v0_8')

df.plot.scatter(x='View Count', 
                y='Total Likes', 
                c='Like/Dislike Ratio', 
                cmap='summer', 
                edgecolor='black',
                linewidth = 1,
                alpha=0.75,
                title='Trending Youtube Videos').get_figure().set_size_inches(13.5, 7)

plt.xscale('log')  # to level out the extreme outliers
plt.yscale('log')

plt.tight_layout()
```
