```python
%matplotlib notebook

import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
from itertools import count
```

```python
plt.style.use('fivethirtyeight')
```

```python
x_vals = []
y_vals = []
```

```python
index = count()
```

```python
def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.cla()  # clears the axis with old lines, so that only new line will be shown
    plt.plot(x_vals, y_vals)
```

```python
ani = FuncAnimation(fig=plt.gcf(), func=animate, interval=1000)

plt.tight_layout()
```
