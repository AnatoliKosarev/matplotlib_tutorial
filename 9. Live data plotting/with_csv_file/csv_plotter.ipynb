```python
%matplotlib notebook

import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
```

```python
plt.style.use('fivethirtyeight')

def animate(i):
    df = pd.read_csv('data.csv')
    
    x = df['x_value']
    
    plt.cla()  # clears old lines, only the new one is shown
    
    plt.plot(x, df['total_1'], label='Channel 1')
    plt.plot(x, df['total_2'], label='Channel 2')
    
    plt.gcf().set_size_inches(9.5, 5)
    plt.legend(loc='upper left')    
    plt.tight_layout()
    
ani = FuncAnimation(fig=plt.gcf(), func=animate)
```
